```python
def check_compilation(code):
    
    try:
        compile(code, '<string>', 'exec')
        return 1
    except SyntaxError:
        print(code)
        return 0
    
lines = code.split('\n')
added_pass = False
compile_line_rewards = 0
l = []
for i, line in enumerate(lines):
    if lines[i].strip().startswith(('for ', 'if ', 'elif', 'else', 'while')):
        # If it's the beginning of a for loop, add 'pass' and check compilation
        line = f'{line} pass'
        added_pass = True
    
    if added_pass:
        result = check_compilation('\n'.join(lines[:i]) + '\n' + line)
    else:
        result = check_compilation('\n'.join(lines[:i + 1]))
    
    if result==1:
        compile_line_rewards += 1
    
    print(f'Line {i + 1} compile reward : {result}')
    print("-----------------------------------")

print("total code compile reward = ", compile_line_rewards/len(lines))
```

Rules while compilation
- if whole program can compile, high reward
- if each line is compiled, we have line wise compile reward
- do we allow partial compile reward? how ? if 1, 2, 3, 6, 10 line can compile and other can't, what should be the reward? may be 1 for these lines and 0 for other non-compilable lines. or 1 for 1, 2, 3 lines and if line 4 doesn't compile then 0 for all others. We can try both variations.
